## 冰箱系统的基本热力模型
![fig 1](energy_flow.jpg)



流经压缩机的制冷剂质量流量：


$$ \dot{m}_{ref} = \frac {\eta_vV_kN}{v_1}  （1） $$
压缩机消耗的电功率：


$$ P_{electric}= \frac {\dot{m}_{ref}(h_{2s}-h_1)}{\eta_{is}}\centerdot \centerdot \centerdot(2)$$
压缩机壳体的发热量：


$$ Q_{comp} = UA_{comp}(T_2-T_{amb}) = \dot{m}_{ref}[(h_{2s} - h_1)(\frac{1}{\eta_{is}}-1)+(h_{2s}-h_2)] \centerdot \centerdot \centerdot(3)$$


冷凝器散热量：


$$ Q_{condenser} = {UA}_{condenser}(T_2-T_{amb})= \dot{m}_{ref}(h_{2}-h_3)\centerdot \centerdot \centerdot(4) $$
蒸发器散热量：


$$ Q_{evap} = UA_{evap}(T_{freezer}-T_{evap})= \dot{m}_{ref}(h_{1}-h_3)\centerdot \centerdot \centerdot(5) $$
毛细管-回气管换热器：


$$ \epsilon_{ihx}= \frac {T_1 - T_5}{T_3-T_4}\centerdot \centerdot \centerdot (6)$$
压缩机运行率：


$$ r_t = \frac {Q_{cab}}{Q_{evap}}\centerdot \centerdot \centerdot (7) $$
制冷系统COP：


$$ COP_{sys} = \frac{h_1-h_3}{h_{2s}-h_1} \centerdot \eta_{is} \centerdot \centerdot \centerdot(8) $$


优化箱体设计，能够降低$Q_{cab}$;优化冷凝器和蒸发器的配置，能够增大$UA_{cond}$和$UA_{evap}$;优化毛细管-回气管换热器的设计，能够增大$\epsilon_{ihx}$ ，降低压缩机的排量和转速，这些手段都有助于让系统获得较低的冷凝温度，较高的蒸发温度，同时获得最高的系统$COP$。

### 热负荷的估计
&emsp;&emsp;冰箱的热负荷是在做制冷系统匹配时最重要的控制环节：箱体设计热负荷偏高，则压缩机选型时排量要增大，COP要增加；蒸发器、冷凝器的设计容量全部要增加，产生的后果是系统匹配难度和设计成本的全面上升. 

&emsp;&emsp;需要特别注意的是，冰箱系统的热负荷随着环温增加是非线性的，其一阶导数是逐渐增加的。即环温越高，热负荷增加的就越快。热负荷的计算可以用下式：
$$ Q_{cab} = \sum_{i=1}^n K_iA_i\delta T_i $$ 
$$\delta T = T_{out}-T_{in}$$
在上述公式中，环温升高，箱体外表面$T_{out}$温度相应升高，内部$T_{in}$维持不变，则传热温差与环温上升是线性关系，但传热K值也会随环温上升而增加，从而使得总漏热量迅速增加。

### 系统工况
&emsp;&emsp;优化的系统工况是保证制冷系统性能和控制成本的基础
从（8）式来看，系统的COP高低由两部分的乘积决定，一部分是制冷系统的热力循环的COP，一部分是压缩机的效率，第一部分$\frac{h_1-h_3}{h_{2s}-h_1}$掌握在我们自己手中，第二部分$\eta_{is}$掌握在我们和压缩机制造商手中。 

&emsp;&emsp;系统工况指在冰箱运行至稳态或近稳态时，冰箱系统管路内制冷剂的实际循环流动的状态，具体来讲，就是冰箱在运行至可判稳状态时，其排气压力（温度）、冷凝压力（温度），过冷温度、蒸发压力（温度）、吸气压力（温度）等。
#### 实际冰箱的系统工况的特点如下：
&emsp;&emsp;由于冰箱制冷系统为小型制冷系统，冷凝温度一般比环境温度高5~10度，蒸发温度比冷冻室空气温度低5~10度，过冷温度一般比冷凝温度低1~3度，回气温度比环温低0~3度；高压侧压力和饱和冷凝压力相同，低压侧压力与饱和蒸发压力相同，由于蒸发器管路内有液态制冷剂流过，其管道阻力较大，到蒸发器末端，由于蒸发压力的下降产生蒸发温度下降的情况。

工况变化对蒸汽压缩循环性能的影响：
>1 蒸发压力不变，冷凝温度升高(近似相当于减少冷凝器设计容量）
回气温度随冷凝温度升高而升高；单位容积制冷量下降；单位容积有效输入功需求变大；系统COP下降
                 
>2 冷凝压力不变，蒸发温度升高（相当于增大蒸发器设计容量）
回气温度随蒸发温度升高而升高；单位容积制冷量上升，单位容积有效输入功需求变小，系统COP升高

工况对压缩机性能的影响：
>1、吸气流量
&emsp;&emsp;吸气密度与吸气压力和吸气温度有关,吸气压力越大，吸气密度越高；吸气温度升高，吸气密度降低
                 $$\rho = \frac{P}{kT}$$
&emsp;&emsp;压缩机容积效率曲线和高低压比的关系如图：当冷凝温度不变，蒸发温度降低，压比增大，则吸气效率降低；当蒸发温度不变，冷凝压力升高，则压比增大，吸气效率降低。
                 
                 
>2、电机效率
                 
工况对压缩机效率的影响是比较复杂的，因为压缩机效率的高低不仅和机械部件的受力情况相关，和电机本身的设计相关性也非常高；
目前定频压缩机采用单相异步电机作为驱动部件，而变频压缩机采用永磁同步电机作为驱动部件，两种电机的特点是截然不同的，对匹配工程师来讲，最重要的了解以下特点：
1、电机有最优效率点，压缩机厂家可以根据产品设计需求进行调整。
2、单相异步电机的效率受吸排气压力影响的变化幅度较高；而变频压缩机使用的同步电机的效率随工况变化的幅度相对较小，但也有可优化空间。
3、一般压缩机厂家以优化标况（ASHRAE工况或CECOMAF工况）作为电机优化设计的目标，该电机最高效率点基本对应冰箱的高环温工况，和冰箱的能耗测试工况相去甚远，所以经常出现标称值很高，能效测试结果很差的局面。


### 压缩机的选型
   根据最高负荷和工况选择压缩机排量：
   $$ Q_e = \dot{m}_{ref}(h_1 - h_3)=\frac{\eta_v V_kN_{max}}{v_1}(h_1-h_3)>Q_{cab(max)}$$
   根据稳态热负荷和稳态工况选择压缩机COP档位
   $$ \frac{Q_{cab}}{COP_{sys}}+Ec_{other} + Ec_{defrost}<Ec_{max}$$
### 换热器设计
    最大限度增加换热器的空气侧有效换热面积，获取最低的换热器热阻
### 毛细管-回气管换热器的设计
    尽量使用低流量毛细管，增大毛细管流量要以冷凝器和蒸发器设计容量富余为前提
    尽可能增大回热系数
    调整毛细管流量或者回热设计后，充注量需要重新匹配。
   
    
    

### 制冷剂在制冷系统中的迁移、分布
&emsp;&emsp;1、冰箱未制冷，内外温度、压力平衡的状态下，制冷剂除部分溶解在压缩机润滑油内，其它全部以气态存在于系统中。<br>
&emsp;&emsp;2、开机后，制冷剂在压缩机驱动下，向冷凝器侧迁移，此时压缩机流量会高于毛细管内的制冷剂流量，此后压缩机流量下降，毛细管流量上升，直至两者相等。<br>
&emsp;&emsp;3、制冷剂的充注量匹配，就是要保证制冷系统在各种状态下，蒸发器维持最低的过热度，同时不会有液态制冷剂通过回气管线进入压缩机内。<br>
而蒸发器的管内容积对制冷剂充注量的多少影响较大，而冷凝器影响较小，但如果冷凝器设计容量过大，过冷段管线太长，则系统在低环温运行时，制冷剂长时间存在于冷凝器内部，造成蒸发器严重缺液，系统蒸发压力过低，流量太小，冰箱将无法正常制冷。<br>
&emsp;&emsp;4、由于开机时，制冷剂需要从蒸发器向冷凝器迁移；停机后，制冷剂又通过毛细管平衡作用向低压侧迁移，这两个过程会额外消耗压缩机的输入功；如果在系统的毛细管入口前增加阀门，压缩机开机时打开，压缩机停机后关闭，则可以迅速实现高低压力平衡、流量平衡；节约部分电能。这一方法在定频冰箱、运行率较低的变频冰箱上，有明显的节能效果。


|状态/部件|压缩机|冷凝器|毛细管|蒸发器|
|:--:|:--:|:--:|:--:|:--:|
|开机初始段|刚开机时压缩机吸气流量最大，此后随吸气压力降低而降低|排入冷凝器的气体增多，压力、温度都升高；当压力上升至气体温度对应的饱和压力之上时，开始有制冷剂液体冷凝并流向毛细管|随着毛细管两端压差突然增加，流量开始增大，但初始状态下毛细管流量远低于压缩机流量；当冷凝器有液态制冷剂流过后，毛细管入口为两相，流量显著增大；|蒸发器内的制冷剂开始朝压缩机吸气侧移动，同时压力和温度降低；如果有残余制冷剂，则一部分蒸发，一部分在储液器内被截流，初始流量最大，随后逐渐降低。|
|稳态运行|流量稳定，制冷剂状态随工况变化而变化|冷凝压力和制冷剂的饱和冷凝压力相同，冷凝器末端形成稳定的过冷区，一般存在1~2度的过冷度，但过高则需要小心。|毛细管内质量流量和压缩机流量相同，随着蒸发温度降低而同步逐渐降低。|蒸发器内制冷剂流量稳定，随着箱体内空气温度降低，蒸发温度降低，流量逐渐降低。|
|压缩机停机|压缩机停转后，压缩机吸排气侧压差将降低；流量瞬间下降至0；|冷凝器内压力迅速降低，而制冷剂温度开始缓慢降低，当制冷剂温度高于压力对应的饱和蒸发温度时，制冷剂开始蒸发，造成冷凝器末端温度陡然下降；残余的制冷剂流向毛细管，流量迅速降低。|停机后残余的两相制冷剂会在残余压差的作用下，继续向低压侧流动，整个过程会持续大概5~7min；当毛细管两侧压力平衡后，流量降低为零。|残余的两相制冷剂流入蒸发器，会造成蒸发器入口段温度上升速度快于蒸发器出口段，这在型式实验时很容易观察到。会有残余的液态制冷剂躲在冷冻蒸发器内，此时蒸发器内的制冷剂质量多于冷凝器内。|

    
    
    

### 制冷系统的瞬时行为

```python
import CoolProp.CoolProp as CP
from RefCycle import refCycle
import numpy as np
import scipy.linalg as alg
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def suctionTemp(Tk,Te,Tsub,eihx):
    return eihx*(Tk-Tsub-Te)+Te

#定义系统循环：

ref = "R600a"
subcoolDegree = 2
condTemp = np.linspace(35,45,11)
evapTemp = np.linspace(-30,-20,11)
Eihx = 0.9

```

```python
#当蒸发温度维持在-25度，冷凝温度从35变化至45度，过冷度2K维持不变
ST = suctionTemp(condTemp,-25,subcoolDegree,Eihx)
plt.figure(figsize=(10, 5))
plt.plot(condTemp, ST)
plt.ylabel('suction Temperature')
plt.show()
Qe = refCycle(ref,ST,-25,condTemp,condTemp-subcoolDegree).qe()
plt.figure(figsize=(10, 5))
plt.plot(condTemp, Qe)
plt.ylabel('Cooling Capacity')
plt.show()

Wc = refCycle(ref,ST,-25,condTemp,condTemp-subcoolDegree).wc()
plt.figure(figsize=(10, 5))
plt.plot(condTemp,Wc)
plt.ylabel('Effective Work Input')
plt.show()
COP =refCycle(ref,ST,-25,condTemp,condTemp-subcoolDegree).cycleCOP()
plt.figure(figsize=(10, 5))
plt.plot(condTemp,COP)
plt.ylabel('Cycle COP')
plt.show()
```

```python
# 冷凝温度不变，40度，蒸发温度从-30~-20
ST = suctionTemp(40,evapTemp,subcoolDegree,Eihx)
plt.figure(figsize=(10, 5))
plt.plot(evapTemp, ST)
plt.ylabel('suction Temperature')
plt.show()
Qe = refCycle(ref,ST,evapTemp,40,40-subcoolDegree).qe()
plt.figure(figsize=(10, 5))
plt.plot(evapTemp, Qe)
plt.ylabel('Cooling Capacity')
plt.show()

Wc = refCycle(ref,ST,evapTemp,40,40-subcoolDegree).wc()
plt.figure(figsize=(10, 5))
plt.plot(evapTemp,Wc)
plt.ylabel('Effective Work Input')
plt.show()
COP =refCycle(ref,ST,evapTemp,40,40-subcoolDegree).cycleCOP()
plt.figure(figsize=(10, 5))
plt.plot(evapTemp,COP)
plt.ylabel('Cycle COP')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(111)
ax2 =ax1.twinx()

ax1.plot(evapTemp,Qe/1000,'r--',evapTemp,Wc/1000,'g--')
ax1.set_ylabel("capacity(KJ/KG)")
ax2.plot(evapTemp,COP,'B',lw=1)
ax2.set_ylabel("CYCLE COP")
ax1.set_xlabel("condensing Temperature(℃)")
plt.show()
```

```python
suctionDensity = refCycle(ref,ST,evapTemp,40,40-subcoolDegree).denSuction() #单位kG/m^3
print(suctionDensity)
plt.figure(figsize=(10, 5))
plt.plot(evapTemp, suctionDensity,"bs")
plt.ylabel('Suction gas density')
plt.show()
```

```python
suct = np.linspace(25,35,11)
suctionDensity = refCycle(ref,suct,-25,40,40-subcoolDegree).denSuction() #单位kG/m^3
print(suctionDensity)
plt.figure(figsize=(10, 5))
plt.plot(suct, suctionDensity,"bs")
plt.ylabel('Suction gas density')
plt.show()
```
