```python
import numpy as np
import matplotlib.pyplot as plt

df=1.5e-5 # aluminium foil thickness 0.015mm
df2 = 1.5e-4
dp = 0.45e-3 # side steel plate thickness 0.4mm
kf= 227 # heat conductivity Wm^-1K^-1
kp = 60 # heat conductivity Wm^-1K^-1
h= 10 # heat transfer coefficient Wm^-2K^-1
m = (h/(kf*df+kp*dp))**0.5
R = np.linspace(10,50,9)*1e-3
uf = np.tanh(m*R)/(m*R)
### new data
m2= (h/(kf*df2+kp*dp))**0.5
uf2 = np.tanh(m2*R)/(m2*R)
plt.plot(R, uf,label = "$\delta_f = 0.015mm$",linestyle=':',marker = "x",color ="blue")
plt.plot(R, uf2,label = "$\delta_f = 0.15mm$",linestyle='--',marker = "d",color ="green")
plt.xlabel("R($m$)")
plt.ylabel("Fin efficiency")
plt.legend()
plt.savefig("pl_fin_eff.png",dpi=300,figsize = 1000)
plt.show()
for i in range(0,len(R)):
    print(R[i],uf[i],uf2[i])




```
